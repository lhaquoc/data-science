### Pyplot API vs OO API
* Pyplot API -> Quickly
* OO API -> Advanced

```python
x = np.linspace(0,10,1000)
plt.plot(x, np.sin(x), color='blue', linestyle='dashed', label='sin(x)');
plt.plot(x, np.cos(x), color='red', label='cos(x)')
plt.title('sin(x) and cos(x)')
plt.xlabel('x')
plt.ylabel('f(x)')
#plt.xlim([0,4])
#plt.ylim([start,end])
plt.legend();
```
