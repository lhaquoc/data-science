### Pyplot API vs OO API
* Pyplot API -> Quickly
* OO API -> Advanced

### Pyplot API

```python
x = np.linspace(0,10,1000)
plt.plot(x, np.sin(x), color='blue', linestyle='dashed', label='sin(x)');
plt.plot(x, np.cos(x), color='red', label='cos(x)')
plt.title('sin(x) and cos(x)')
plt.xlabel('x')
plt.ylabel('f(x)')
#plt.xlim([0,4])
#plt.ylim([start,end])
xmin = 0
xmax = 4
ymin = -0.7
ymax = 0.6
#plt.axis([xmin, xmax, ymin, ymax])
plt.axis('equal')
plt.legend();
```

## OO API

```python
#1 Prepare data
x = [1,2,3,4]
y = [11,22,33,44]

#2 Setup plot
fig, ax = plt.subplots(figsize=(5,5)) #Figure size = W and H of the plot

#3 Plot the data
ax.plot(x, y);

ax.set(title='A simple plot', xlabel='x-axis', ylabel='y-label')
```

## Most common types of Matplotlib plots

* scatter
* bar

```python
# Create an array
x = np.linspace(0,10,1000)
x[:5]
```

```python
# OO API plots a line chart
fig, ax = plt.subplots()
ax.plot(x, x**3);
```

## Scatter

```python
# Pyplot API
plt.scatter(x, np.exp(x)); # y = e^x
```

```python
# OO API
#1 Prepare data
rng = np.random.RandomState(0)
x = rng.randn(100)
y = rng.randn(100)
colors = rng.randn(100)  
sizes = 1000*rng.rand(100)
#2 Setup plot
fig, ax = plt.subplots()

#3 Plot the data
plot = ax.scatter(x, y, s=sizes, c=colors, cmap='viridis', alpha=0.3);
fig.colorbar(plot);
```

### Bar chart
* Vertical
* Horizontal

```python
# prepare data from dict
soft_drink_price = {'Cocacola': 10,
                    'Pepsi':12,
                    'Sprite':8}

fig, ax = plt.subplots();
ax.bar(soft_drink_price.keys(), soft_drink_price.values())
ax.set(title='Supermarket\'s soft drink prices', ylabel='Price($)');
```

```python
fig, ax = plt.subplots();
ax.barh(list(soft_drink_price.keys()), soft_drink_price.values())
ax.set(title='Supermarket\'s soft drink prices', xlabel='Price($)');
```

## Histogram

```python
# Prepare data
np.random.seed(42)
student_height = np.random.normal(170, 10, 250)
```

```python
student_height[:10]
```

```python
# Pyplot API
plt.hist(student_height, bins=30);
```

## Subplots

```python
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2,
                       ncols=2,
                       figsize=(10,5));
ax1.plot(x, x/2);
ax1.plot(x, x*2);

ax2.scatter(np.random.random(10), np.random.random(10))

ax3.bar(soft_drink_price.keys(), soft_drink_price.values());

ax4.hist(student_height, bins=30);
```

```python
fig, ax = plt.subplots(nrows=2,
                       ncols=2,
                       figsize=(10,5))
ax[0,0].plot(x, x/2);
ax[0,0].plot(x, x*2);

ax[0,1].scatter(np.random.random(10), np.random.random(10))

ax[1,0].bar(soft_drink_price.keys(), soft_drink_price.values());

ax[1,1].hist(student_height, bins=30);
```
